```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method to create a basic plot 
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

### Putting multi-plots on the same canvas

```python
plt.subplot(1,2,1) #Row, Columns, # plot you are reffering to
plt.plot(x,y,'r') # Color Red 'r'

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # Left, Bottom, Width and Height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # These are percent between 0-1
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # Left, Bottom, Width and Height

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3) # Tuple unpacking

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
# since we set axes = plt.subplots axes is really just a list now, and we can treat it like a list by indexing 
fig,axes = plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,y)
axes[1].set_title('Second Plot')

plt.tight_layout() # Fix the overlap
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2), dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,y)

plt.tight_layout()
```

### Save a figure to a file

```python
fig.savefig('my_testfig.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,x **2, label='X Squared')
ax.plot(x,x ** 3, label='X Cubed')
ax.plot(x,x, label='X')

ax.legend(loc=0)

ax.set_title('Title')
ax.set_ylabel('YLabel')
ax.set_xlabel('XLabel')

plt.xlim([0, max(x) + 1])
plt.ylim([0, max(y) + 1])

```
